```python
df = df.rename(columns= {'Cement (component 1)(kg in a m^3 mixture)': 'cement',
                      'Blast Furnace Slag (component 2)(kg in a m^3 mixture)':'slag',
                      'Fly Ash (component 3)(kg in a m^3 mixture)':'Ash',
                      'Water  (component 4)(kg in a m^3 mixture)': 'water',
                      'Superplasticizer (component 5)(kg in a m^3 mixture)':'Superplasticizer',
                      'Coarse Aggregate  (component 6)(kg in a m^3 mixture)':'coarse_aggregate',
                      'Fine Aggregate (component 7)(kg in a m^3 mixture)':'fine_aggregate',
                      'Age (day)':'age',
                      'Concrete compressive strength(MPa, megapascals) ':'strength' })
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(),annot = True)
```

##### There are'nt any high correlations between Compressive strength and other features except for Cement, which should be the case for more strength.Age and Super plasticizer are the other two features which are strongly correlated with Compressive Strength.Super Plasticizer seems to have a negative high correlation with Water, positive correlations with Fly ash and Fine aggregate.

```python
X = df.drop("strength",axis=1)
y = df['strength']
```

```python
#preprocessing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state=51)
```

## Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
#Predicting the Test set results
from sklearn.metrics import r2_score
y_pred_lr = lr.predict(X_train)
score = r2_score(y_train,y_pred_lr)
print("Score of Training:",100*score)
```

```python
#Predicting the Test set results
from sklearn.metrics import r2_score
y_test_pred_lr = lr.predict(X_test)
score = r2_score(y_test,y_test_pred_lr)
print("Score of Testing:",100*score)
```

```python
import xgboost as xg 
xg_reg = xg.XGBRegressor()
xg_reg.fit(X_train,y_train)
```

```python
#Predicting the Train set results

y_pred_xg_reg = xg_reg.predict(X_train)
score = r2_score(y_train,y_pred_xg_reg)
print("Score of Training:",100*score)
```

```python
#Predicting the Test set results
from sklearn.metrics import r2_score
y_test_pred_xg_reg = xg_reg.predict(X_test)
score = r2_score(y_test,y_test_pred_xg_reg)
print("Score of Testing:",100*score)
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold,RandomizedSearchCV
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=800)
rf.fit(X_train, y_train);
```

```python
#Predicting the Test set results

y_pred_rf = rf.predict(X_train)
score = r2_score(y_train,y_pred_rf)
print("Score of Training:",100*score)
```

```python
#Predicting the Test set results
from sklearn.metrics import r2_score
y_test_pred_rf = rf.predict(X_test)
score = r2_score(y_test,y_test_pred_rf)
print("Score of Testing:",100*score)
```

```python
from lightgbm import LGBMRegressor

params={ "n_estimators" : [100,300,500,1000,1500],
       "learning_rate": [0.01,0.02,0.05,0.08]
      }

classifier=LGBMRegressor(random_state=45)

random_search=RandomizedSearchCV(classifier,param_distributions=params,n_iter
=5,scoring='neg_root_mean_squared_error',n_jobs=-1,cv=5,verbose=3)
random_search.fit(X,y)

print('Best rmse: {:.4}, with best C: {}'.format(random_search.best_score_, random_search.best_params_))
```

```python
from sklearn.metrics import mean_squared_error

params = random_search.best_params_

from lightgbm import LGBMRegressor
lgb =  LGBMRegressor(**params,random_state=45)
lgb.fit(X_train,y_train)

y_test_pred_lgb = lgb.predict(X_test)
print("RMSE : " , np.sqrt(mean_squared_error(y_test,y_test_pred_lgb)))
```

```python
y_pred_lg = lgb.predict(X_train)
score = r2_score(y_train,y_pred_lg)
print("Score of Training:",100*score)
```

```python
#Predicting the Test set results
from sklearn.metrics import r2_score
y_test_pred_lg = lgb.predict(X_test)
score = r2_score(y_test,y_test_pred_lg)
print("Score of Testing:",100*score)
```
